```python
# https://www.kaggle.com/residentmario/ramen-ratings
import pandas as pd
import seaborn as sns
```

## Tasks

```python
# 1
data = pd.read_csv('ramen-ratings.csv')
data.head()
```

```python
# 2
data['CountryMeanRating'] = data.groupby('Country')['Stars'].transform(func='mean')
data.head()
```

```python
# 3
data.groupby(['Country','Style']).mean().unstack('Style')[['Stars']]
```

```python
# 4
def shorten_string(row):
    words = row.Variety.split(' ')
    words = words[:2]
    words.append('...')
    return ' '.join(words)
```

```python
# 5
data['ShortVariety'] = data.apply(shorten_string, axis=1)
data.head()
```

```python
# 6
data.set_index('Brand', inplace=True)
data.head()
```
