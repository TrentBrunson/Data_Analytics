```python
# https://www.kaggle.com/neuromusic/avocado-prices
import pandas as pd
```

## Tasks

```python
# 1
data = pd.read_csv('avocado.csv')
data.head()
```

```python
# 2

# convert the columns to a list
allCols = data.columns.tolist()

# create a new list with the formatted column names
newCols = []
for col in allCols:
    c = ''.join(col.title().split(' '))
    newCols.append(c)
    
# assign the list of columns to the columns attribute
data.columns = newCols
    
# rename the Averageprice column to AveragePrice
data.rename(columns={'Averageprice':'AveragePrice'}, inplace=True)

# check that the columns have been renamed
data.info()
```

```python
# 3
data['XLargePercent'] = data['XlargeBags'] / data['TotalBags']
```

```python
# 4
data['LargePercent'] = data['LargeBags'] / data['TotalBags']
```

```python
# 5
data['SmallPercent'] = data['SmallBags'] / data['TotalBags']
```

```python
# 6
data.head()
```

```python
# 7
dataReduced = data[['Region','Type','Year','TotalBags']]
```

```python
# 8
dataGrouped = dataReduced.groupby(['Region','Type','Year']).sum().unstack('Type')
dataGrouped
```
