## Get the data - Create a DataFrame

- A DataFrame (or a DatFrame object) is a Pandas object that stores the data for an analysis. (*Python for Data Analysis* textbook, page 46)
- Pandas DataFrame is two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns). A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns. Pandas DataFrame consists of three principal components, the data, rows, and columns. ("Python Pandas DataFrame," *Geeks for Geeks*, https://www.geeksforgeeks.org/python-pandas-dataframe/)

Three read() methods for importing data into a DataFrame
- read_csv()
- read_excel()
- read_sql_query()

### Another Example of Building a DataFrame Object

```python
with open('justices.txt', 'r') as infile:
    justiceList = [line.rstrip() for line in infile]
#print(justiceList)
for i in range(len(justiceList)):
    justiceList[i] = justiceList[i].split(',')
    justiceList[i][4] = int(justiceList[i][4])
    justiceList[i][5] = int(justiceList[i][5])
    # set 'Year Left Court' for current justices
    if justiceList[i][5] == 0:
        justiceList[i][5] = 2021
justiceList
```

```python
justice_df_data = justiceList
justice_df_columns = ['First Name','Last Name','Appointed By','Home State','Year Appointed','Year Left Court']
```

```python
justices_data = pd.DataFrame(data=justice_df_data, columns=justice_df_columns)
```

```python
justices_data
```

## Shape the data

```python
mortality_data = mortality_data.set_index(
    ['Year','AgeGroup'], verify_integrity=True)
mortality_data.head(2)
```

```python
mortality_data.reset_index(inplace=True)
mortality_data.head(2)
```

### Pivot the data

```python
mortality_wide = mortality_data.pivot(
    index='Year', columns='AgeGroup', values='DeathRate')
mortality_wide.head(3)
```

```python
mortality_wide = mortality_data.pivot(
    index='Year', columns='AgeGroup')
mortality_wide.head(3)
```

### Melt the data

```python
# get starting data
mortality_wide = mortality_data.pivot(
    index='Year', columns='AgeGroup', values='DeathRate')

# save to Excel format to remove indexes
mortality_wide.to_excel('mortality_wide.xlsx')
mortality_wide = pd.read_excel('mortality_wide.xlsx')

mortality_wide.head(4)
```

```python
mortality_long = mortality_wide.melt(
    id_vars = 'Year',
    value_vars=['01-04 Years','05-09 Years'],
    var_name ='AgeGroup',
    value_name='DeathRate')

with pd.option_context('display.max_rows', 4):
    display(mortality_long)
```
